# WEDSCRAPING

```python
! pip install bs4

import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup

import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
import seaborn as sns

# importing modules which are heplful for my project


Company_Name =[]
Company_Business =[]
Company_Ceo =[]
World_Rank_2022 =[]
Market_Value_2022 =[]
World_Rank_2021 =[]
Market_Value_2021 =[]
World_Rank_2020 =[]
Market_Value_2020 =[]
Annual_revenue_in_usd =[]
Annual_net_income_in_Usd=[]
Stock_Exchange =[]


for i in range (0,70):
    
    
    
    url =('https://www.value.today/?title=&field_company_category_primary_target_id&field_headquarters_of_company_target_id=All&field_company_website_uri=&field_market_value_jan072022_value=&page={0}'.format(i))
    
    response = requests.get(url)
    soup = BeautifulSoup(response.text,'html.parser')
    
    CompanyData = soup.findAll('li',attrs ={'row well clearfix'})
    
    
    for Data in CompanyData:
        name = Data.find('div', attrs={'field field--name-node-title field--type-ds field--label-hidden field--item'})
        if name is None:
            Company_Name.append(np.NaN)
        else:
            Company_Name.append(name.text.replace("\n",""))
            
           
        
        business = Data.find('div', attrs={'clearfix col-sm-12 category-button field field--name-field-company-category-primary field--type-entity-reference field--label-above'})
        if business is None:
            Company_Business.append(np.NaN)
        else:
            Company_Business.append(business.text.replace('\n',"").replace("Company Business",""))
            
       
        
        Ceo = Data.find('div',attrs={'clearfix col-sm-12 field field--name-field-ceo field--type-entity-reference field--label-above'})
        
        if Ceo is None:
            Company_Ceo.append(np.NaN)
        else :
            Company_Ceo.append(Ceo.text.replace("\nCEO:\n","").replace("\n",""))
            
            
    
        Rank_2022 = Data.find('div',attrs={'clearfix col-sm-6 field field--name-field-world-rank-jan072022 field--type-integer field--label-above'})
        if Rank_2022 is None:
            World_Rank_2022.append(np.NaN)
        else:
            World_Rank_2022.append(Rank_2022.text.replace(',',"").replace("\n","").replace('World Rank (Jan-07-2022)',""))
            
            
            
        Mv_2022 = Data.find('div',attrs={'clearfix col-sm-6 field field--name-field-market-value-jan072022 field--type-float field--label-above'})
        if Mv_2022 is None :
            Market_Value_2022.append(np.NaN)
        else :
            Market_Value_2022.append(Mv_2022.text.replace(',',"").replace("Billion USD","").replace("\n","").replace('Market Value (Jan-07-2022)',"") )
            
            
        
        Rank_2021 = Data.find('div',attrs={'clearfix col-sm-6 field field--name-field-world-rank-jan012021 field--type-integer field--label-above'})
        if Rank_2021 is None:
            World_Rank_2021.append(np.NaN)
        else:
            World_Rank_2021.append(Rank_2021.text.replace(',',"").replace("\n","").replace('World Rank (Jan-01-2021)',""))
            
            
        
        Mv_2021 = Data.find('div',attrs={'clearfix col-sm-6 field field--name-field-market-value-jan012021 field--type-float field--label-above'})
        if  Mv_2021 is None:
            Market_Value_2021.append(np.NaN)
        else :
            Market_Value_2021.append(Mv_2021.text.replace(',',"").replace("Billion USD","").replace("\n","").replace('Market Value (Jan-01-2021)',""))
            

        
        Rank_2020 = Data.find('div',attrs={'clearfix col-sm-6 field field--name-field-world-rank-jan-2020 field--type-integer field--label-above'})
        if Rank_2020 is None:
            World_Rank_2020.append(np.NaN)
        else:
            World_Rank_2020.append(Rank_2020.text.replace(',',"").replace("\n","").replace('World Rank (Jan-2020)',""))
            

            
        Mv_2020 = Data.find('div',attrs={'clearfix col-sm-6 field field--name-field-market-value-jan-2020 field--type-float field--label-above'})
        if  Mv_2020 is None:
            Market_Value_2020.append(np.NaN)
        else:
            Market_Value_2020.append(Mv_2020.text.replace(',',"").replace("Billion USD","").replace("\n","").replace('Market Value (Jan 1st 2020)',""))
            
     
    
        revenue = Data.find('div',attrs={'clearfix col-sm-12 field field--name-field-revenue-in-usd field--type-float field--label-inline'})
        if revenue is None:
            Annual_revenue_in_usd.append(np.NaN)
        else:
            Annual_revenue_in_usd.append(revenue.text.replace("Million USD","").replace(',',"").replace("\n","").replace("Annual Revenue in USD",""))
      
    
    
        net_income = Data.find('div',attrs={'clearfix col-sm-12 field field--name-field-net-income-in-usd field--type-float field--label-inline'})
        if net_income is None:
            Annual_net_income_in_Usd.append(np.NaN)
        else:
            Annual_net_income_in_Usd.append(net_income.text.replace("Million USD","").replace(',',"").replace("\n","").replace("Annual Net Income in USD",""))
            
            
   
        Exchange = Data.find('div',attrs={'clearfix col-sm-12 field field--name-field-stock-exchange-lc field--type-entity-reference field--label-inline'})
        if Exchange is None:
            Stock_Exchange.append(np.NaN)
            
        else:
            Stock_Exchange.append(Exchange.text.replace("\n","").replace("Stock Exchange","").replace("NasdaqBovespa Brazil","Nasdaq Bovespa Brazil").replace("NYSEEURONEXT","NYSE EURONEXT").replace("NYSEBovespa Brazil","NYSE Bovespa Brazil").replace("NYSESwitzerland","NYSE Switzerland").replace("NYSEBovespa BrazilSwitzerlandEURONEXT","NYSE Bovespa Brazil,Switzerland,EURONEXT").replace("NYSE Bovespa BrazilSwitzerlandEURONEXT","NYSE Bovespa Brazil,Switzerland,EURONEXT").replace("NYSELSE","NYSE LSE").replace("NYSEFrankfurt","NYSE Frankfurt").replace("Deutsche BorseNYSE","Deutsche Borse,NYSE").replace("NasdaqMexico","Nasdaq Mexico").replace("NYSEMexico","NYSE Mexico").replace("NYSE EURONEXTLSESwitzerland","NYSE EURONEXT,LSE,Switzerland"))
            
            
    
    
    # NaN stands for Not Available
    # html code is converted to text 
    # unwanted text/symbols are replaced with empty space 
```
